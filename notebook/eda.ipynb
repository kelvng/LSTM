```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15.0, 5.0)
```

```python
train = pd.read_csv('../../data/word_seperate/corpus', header=None)
```

```python
words = ' '.join(train[1].values.tolist()).split(' ')
counts = pd.value_counts(words)
```

```python
counts.head(30).plot.bar()
```

```python
np.savetxt('word_vocab', counts.head(5000).index.values, delimiter='\n', fmt='%s')
```

```python
chars = list(''.join(train[0].values.tolist()))
```

## write down char vocab

```python
char_counts = pd.value_counts(chars)
np.savetxt('vocab', char_counts.head(108).index.values, delimiter='\n', fmt='%s')
```
